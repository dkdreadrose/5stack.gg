```python
# get puuid for each player
summoners = [summoner for summoner in summoners if summoner['riotId'] is not None and summoner['riotId'] != '']
for summoner in summoners:
    riotId = summoner['riotId']
    game_name = '#'.join(riotId.split('#')[:-1])
    tag_line = riotId.split('#')[-1]
    summoner['puuid'] = riot_watcher.account.by_riot_id(summoner['riot_region'], game_name, tag_line)['puuid']

for summoner in summoners:
    print(summoner)
```

```python
team_matches = []
# get all matches for player 1, then filter for matches all other players were in
match_ids = lol_watcher.match.matchlist_by_puuid(region=summoners[0]['riot_region'], puuid=summoners[0]['puuid'])
print(match_ids)
for match_id in match_ids:
    match = lol_watcher.match.by_id(region=summoners[0]['riot_region'], match_id=match_id)
    if all(summoner['puuid'] in match['metadata']['participants'] for summoner in summoners[1:]):
        match_timeline = lol_watcher.match.timeline_by_match(region= summoners[0]['riot_region'], match_id=match_id)
        match['timeline'] = match_timeline
        team_matches.append(match)
print(len(team_matches))
```
